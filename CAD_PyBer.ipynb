```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
df = pd.merge(ride_data, city_data, how="left", on="city")
# Display the data table for preview
df.head()
```

## Bubble Plot of Ride Sharing Data

```python
max(df["fare"].value_counts().sort_index())
```

```python
# Obtain the x and y coordinates for each of the three city types
city_groups = df.groupby(["city"])

x_values = df["city"].value_counts().sort_index()
y_values = city_groups["fare"].mean()

# Build the scatter plots for each city types
colors = {
    "Urban": "lightcoral",
    "Suburban": "lightskyblue",
    "Rural": "gold"
}

# Incorporate the other graph properties
sizes = city_groups["driver_count"].first()*10
# linewith
# edgecolor
# Create a legend

# legend = plt.legend(fontsize = 8, title= "City Types", loc="best") 
# legend.legendHandles[1]._sizes = [10]
# legend.legendHandles[2]._sizes = [10]
# legend.legendHandles[3]._sizes = [10]

plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
# Incorporate a text label regarding circle size
plt.text(42,20,": \nCircle size correlates with driver count per city.", fontsize = 8)
# Save Figure
plt.scatter(x_values, y_values, s=sizes, c=[ colors[city_type] for city_type in city_groups["type"].first() ], linewidth=1, edgecolor="black", alpha=0.75)

plt.grid()
```

```python
# Show plot
plt.show()
```
